# Outlier Detection and Analysis

```python
#Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

#STEP 1: Select numeric columns to check for outliers
numeric_cols = ['age', 'absences', 'G1', 'G2', 'G3', 'failures']

#STEP 2: Plot all boxplots together (Before Cleaning)
plt.figure(figsize=(10, 6))  # control figure size
sns.boxplot(data=student_pd[numeric_cols])
plt.title("Boxplots Before Cleaning")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.show()

#STEP 3: Remove unrealistic or extreme values
student_pd = student_pd[student_pd['age'] <= 22]          # most students are under 22
student_pd = student_pd[student_pd['absences'] < 100]     # remove extreme absences
student_pd = student_pd[student_pd['G1'] <= 20]
student_pd = student_pd[student_pd['G2'] <= 20]
student_pd = student_pd[student_pd['G3'] <= 20]
student_pd = student_pd[student_pd['failures'] <= 3]

#STEP 4: Plot all boxplots together (After Cleaning)
plt.figure(figsize=(10, 6))
sns.boxplot(data=student_pd[numeric_cols])
plt.title("Boxplots After Cleaning")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.show()
```

```python
# Import LabelEncoder from scikit-learn
from sklearn.preprocessing import LabelEncoder

#Step 1: Select categorical columns to encode with LabelEncoder
categorical_cols_label_encode = ['sex', 'address',
                                 'schoolsup', 'famsup', 'paid',
                                 'activities', 'nursery', 'higher', 'internet', 'romantic', 'famsize', 'school', 'Pstatus']

# Step 2: Initialize LabelEncoder
le = LabelEncoder()

#Step 3: Apply Label Encoding to each selected categorical column
for col in categorical_cols_label_encode:
    # Check if the column exists before encoding
    if col in student_pd.columns:
        student_pd[col] = le.fit_transform(student_pd[col])  # Convert categories to numbers
    else:
        print(f"Column '{col}' not found in the DataFrame for Label Encoding.")

# Display the head of the DataFrame after Label Encoding
display(student_pd.head())
```

```python
# Apply one-hot encoding to the specified categorical columns
categorical_cols_one_hot = ['Mjob', 'Fjob', 'guardian', 'reason']

# Check if the columns exist before applying get_dummies
cols_to_one_hot = [col for col in categorical_cols_one_hot if col in student_pd.columns]

if cols_to_one_hot:
    student_pd = pd.get_dummies(student_pd, columns=cols_to_one_hot, drop_first=True)
else:
    print("None of the specified columns for one-hot encoding were found in the DataFrame.")

# Display the head of the DataFrame after one-hot encoding
display(student_pd.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)

from sklearn.tree import DecisionTreeRegressor
decision_tree = DecisionTreeRegressor()
decision_tree.fit(X_train_scaled, y_train)

from sklearn.ensemble import RandomForestRegressor , GradientBoostingRegressor
random_forest = RandomForestRegressor()
random_forest.fit(X_train_scaled, y_train)

gradient_boost = GradientBoostingRegressor()
gradient_boost.fit(X_train_scaled, y_train)

models = {
    'Linear Regression': linear_model,
    'Decision Tree': decision_tree,
    'Random Forest': random_forest,
    'Gradient Boosting': gradient_boost
}

for name, model in models.items():
    print(f"Model Loaded: {name}")
```

```python
from sklearn.metrics import r2_score
y_pred = linear_model.predict(X_test_scaled)
print(f"R2 Score of linear model: {r2_score(y_test, y_pred)}")

y_pred = decision_tree.predict(X_test_scaled)
print(f"R2 Score of Decision tree: {r2_score(y_test, y_pred)}")

y_pred = random_forest.predict(X_test_scaled)
print(f"R2 Score of Random forest: {r2_score(y_test, y_pred)}")

y_pred = gradient_boost.predict(X_test_scaled)
print(f"R2 Score of Gradient boosting: {r2_score(y_test, y_pred)}")
```

```python
from sklearn.metrics import mean_absolute_error

y_pred = linear_model.predict(X_test_scaled)
print(f"MAE of linear model: {mean_absolute_error(y_test, y_pred)}")

y_pred = decision_tree.predict(X_test_scaled)
print(f"MAE of Decision tree: {mean_absolute_error(y_test, y_pred)}")

y_pred = random_forest.predict(X_test_scaled)
print(f"MAE of Random forest: {mean_absolute_error(y_test, y_pred)}")

y_pred = gradient_boost.predict(X_test_scaled)
print(f"MAE of Gradient boosting: {mean_absolute_error(y_test, y_pred)}")
```
